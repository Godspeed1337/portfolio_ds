### Проверка формулы вычисления эффективности обогащения

```python
def recovery(row):
    c = row['rougher.output.concentrate_au']
    f = row['rougher.input.feed_au']
    t = row['rougher.output.tail_au']
    return (c * (f - t)) / (f * (c - t)) * 100
```

```python
train_data['recovery_check'] = train_data.apply(recovery, axis=1)
```

```python
print(mean_absolute_error(train_data['rougher.output.recovery'], train_data['recovery_check']))
```

```python
def get_best_corr(data: pd.DataFrame):
    columns = data.columns.to_list()
    columns.remove('date')
    columns_na = data.columns[data.isna().any()].tolist()
    columns_dir = {}
    for column in columns_na:
        best_corr = -1
        best_column = ''
        for column_2 in columns:
            corr = data[column].corr(data[column_2])
            if corr > best_corr and column != column_2 and corr > 0.3:
                best_corr = corr
                best_column = column_2
        if best_corr != -1:
            print('Na column -', column, 'corr :', best_corr, best_column)
            columns_dir[column] = best_column
    return columns_dir
```

```python
columns_dir = get_best_corr(train_data)
```

```python
def fill_na_corr(data: pd.DataFrame, columns_dir):
    for column_na, column_2 in columns_dir.items():
        categories = pd.qcut(data[column_2], 6)
        data['cat'] = categories
        data[column_na] = data[column_na].fillna(data.groupby('cat')[column_na].transform('median'))
```

### Функция для вычисления итогового sMAPE

Нам нужно написать эту функцию, а в конце сделать это:

```python
def sMAPE(y_target, y_pred):
    return ((1/len(y_target)) * np.sum(2 * np.abs(y_target - y_pred) / (np.abs(y_target) + np.abs(y_pred)))) * 100
```

```python
def final_sMAPE(smape1, smape2):
    return 0.25 * smape1 + 0.75 * smape2
```

```python
sMAPE_score = make_scorer(sMAPE, greater_is_better=False)
```
